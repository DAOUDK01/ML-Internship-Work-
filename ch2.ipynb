## Categorizing Data into three category for NumStorePurchase in marketing_data datasets

```python
marketing_data['bins'] = pd.cut(x=marketing_data['NumStorePurchases'], bins=3, labels=['Low', 'Moderate', 'High'])
display(marketing_data[['NumStorePurchases', 'bins']].head())
```

 the 'NumStorePurchases' values and the assigned categories ('Low', 'Moderate', or 'High') based on which bin each value falls into.
